# Semantic Categories

Using prompt written by the LLM in main3.ipynb.

```typescript
import { START, END, StateGraph, MemorySaver } from "@langchain/langgraph";
import { SystemMessage, HumanMessage } from "@langchain/core/messages";
import { ChatPromptTemplate, MessagesPlaceholder } from "@langchain/core/prompts";
import { JsonOutputParser } from "@langchain/core/output_parsers";
import { readFileSync } from 'node:fs';

import { EXPERIMENTS_DIR, SERVER_DATA_DIR } from '../server/src/util/fileUtils.ts';
import { getNotebookLogger } from '../server/src/Logger.ts';
import { newModel, StateInfo, responseContent } from '../server/src/agents/agent.ts';
import { modelNode } from '../server/src/agents/nodes/modelNode.ts';

// Note that the chain filters out non-JSON parts of the response.
// See that I injected a request for explanations in the prompt.
const prompt = ChatPromptTemplate.fromMessages([ new MessagesPlaceholder("messages") ]);
const llm = newModel("Anthropic");
const parser = new JsonOutputParser();
const chain = prompt.pipe(llm).pipe(parser);

const logger = getNotebookLogger();
const userUUID: string = "0";
const config = { configurable: { thread_id: userUUID } };
const lhsText = readFileSync(`${SERVER_DATA_DIR}/SHA-1-md/selected-text.txt`, 'utf-8');
const PROMPT = readFileSync(`${EXPERIMENTS_DIR}/annotateNodePromptCategories4.txt`, 'utf-8')+"\nPlease also provide some explanation with the JSON array in your output.\n";

var userInput = lhsText;
var output = await chain.invoke({ messages: [
  new SystemMessage(PROMPT),
  new HumanMessage(userInput)
]});
logger.info(output);
```
